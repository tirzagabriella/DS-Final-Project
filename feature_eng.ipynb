```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import category_encoders as ce
```

```python
# Specify the list of categorical columns to be binary encoded
categorical_columns = ['service', 'source', 'country', 'adnet', 'browser', 'profile_status', 'publisher']

# Initialize the binary encoder
encoder = ce.BinaryEncoder(cols=categorical_columns)

# Perform binary encoding on the selected columns
df_encoded = encoder.fit_transform(df)
```
